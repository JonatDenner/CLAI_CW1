## We can see there's no null values or non-numeric data, so plotting the correlation matrix for the fraudulent transactions

```python
df1 = df.copy()
df1 = df1[df.Class != 0]
corr = df1.corr()

# create mask to hide half the plot
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(16, 16))

# Generate a custom diverging colormap
cmap = sns.color_palette("vlag_r", as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap,  vmin=-1, vmax=1, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})


# time has no correlation
df = df.drop("Time", axis=1)
```

```python
x = df["V1"]
y = df["V3"]

fig = plt.figure(figsize = (16, 9))

my_cmap = plt.get_cmap('jet_r')

sctt = plt.scatter(x, y, c=df["Class"], alpha=0.5, cmap = my_cmap)
fig.colorbar(sctt, shrink = 0.5, aspect = 5, label = "Class")

plt.title("V1 Plotted against V3")
 
# show plot
plt.show()
```

# Undersampling
I found that oversampling led to too many data points, making training time unreasonably long with most models.

```python
# Separate fraudulent and non-fraudulent cases
fraud = df[df.Class != 0]
not_fraud = df[df.Class == 0]

# Generate an even spread of rows to take from the non-fraudulent cases
# np linspace gets an even spread of numbers between two numbers
tmp = not_fraud.iloc[np.linspace(0, not_fraud.shape[0]-1, num=fraud.shape[0], dtype=int)].copy()

print("Number of fraudulent entries: {}\nNumber of non-fraudulent entries: {}".format(fraud.shape[0], tmp.shape[0]))
```

We see that there's the same number of fraudulent and non fraudulent cases now

```python
# create test and train sets
from sklearn.model_selection import train_test_split

X = pd.concat([fraud, tmp])

fig = plt.figure(figsize = (16, 9))
sctt2 = plt.scatter(X["V1"], X["V3"], c=X["Class"], alpha=0.5, cmap = my_cmap)
fig.colorbar(sctt2, shrink = 0.5, aspect = 5, label = "Class")
plt.title("V1 Plotted against V3")
plt.show()

y = X["Class"].copy()
X = X.drop("Class", axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)


```

```python
# Try various models for accuracy
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

from sklearn.metrics import accuracy_score

models = []

NaiveBayes = GaussianNB()
SVM = SVC()
Forest = RandomForestClassifier()
KNC = KNeighborsClassifier()
Adaboost = AdaBoostClassifier()
NeuralNet = MLPClassifier()
LDA = LinearDiscriminantAnalysis()

models.append(NaiveBayes)
models.append(SVM)
models.append(Forest)
models.append(KNC)
models.append(Adaboost)
models.append(NeuralNet)
models.append(LDA)

for x in models:
    x.fit(X_train, y_train)
    y_pred = x.predict(X_test)
    print("Model/Accuracy: " + str(x).replace('()', '') + "/%.2f" % accuracy_score(y_test, y_pred) + "%")
    
```
